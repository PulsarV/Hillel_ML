```python
#Imports
import numpy as np
```

```python
# 1. Знайти в датасеті таргет. Завантажити датасет без таргета

source_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
data = np.loadtxt(source_url, delimiter = ',', usecols=(0,1,2,3), dtype=float)
```

```python
# 2. Перетворити колонки, що залишились в 2D масив (або впевнитись, що це уже 2D масив)
print(f"Data ndim = {data.ndim}")
print(f"Data shape = {data.shape}")
print(f"Data sample = {data[0]}")
```

```python
# 3. Порахувати mean, median, standard deviation для 1-ї колонки

first_col = data[:, 0]
print(f"First column mean = {np.mean(first_col)}")
print(f"First column median = {np.median(first_col)}")
print(f"First column standard deviation = {np.std(first_col)}")
```

```python
# 4. Вставити 20 значень np.nan на випадкові позиції в масиві (при використанні звичайного рандому можуть накластись позиції, тому знайти рішення, яке гарантує 20 унікальних позицій)

indices = np.random.choice(data.size, size=20, replace=False)
data.flat[indices] = np.nan
print(f"Data\n{data}")
```

```python
# 5. Знайти позиції вставлених значень np.nan в 1-й колонці

nan_positions = np.where(np.isnan(first_col))[0]
print(f"NaN rows = {nan_positions}")
```

```python
# 6. Замінити всі значення np.nan на 0

data[np.isnan(data)] = 0
print(f"Data\n{data}")
```

```python
# 7. Відфільтрувати массив за умовою: значення в 3-й колонці > 1.5 та значения в 1-й колонці < 5.0 (зберегти у іншу змінну)

mask = (data[:, 2] > 1.5) & (data[:, 0] < 5.0)
filtered = data[mask]
print(f"Filtered data\n{filtered}")
```

```python
# 8. Розбити масив по вертикалі на 2 рівні частини (не використовувати абсолютні числа, мають бути два массиви по 4 колонки)

top, bottom = np.vsplit(data, 2)
print(f"Top data\n{top}")
print(f"Bottom data\n{bottom}")
```

```python
# 9. Відсортувати обидва массиви по 1-й колонці: 1-й за збільшенням, 2-й за зменшенням

top_sorted = top[np.argsort(top[:, 0])]
print(f"Top sorted data\n{top_sorted}")
bottom_sorted = bottom[np.argsort(-bottom[:, 0])]
print(f"Bottom sorted data\n{bottom_sorted}")
```

```python
# 10. Зібрати обидва массиви в одне ціле

data = np.vstack([top_sorted, bottom_sorted])
print(f"New data\n{data}")
```

```python
# 11. Порахувати всі унікальні значення в массиві та вивести їх разом із кількістю

values, counts = np.unique(data, return_counts=True)
print(f"Unique values\n{np.column_stack((values, counts))}")
```

```python
# 12. Знайти найбільш часто повторюване значення в массиві

most_common = values[np.argmax(counts)]
print(f"Most common value = {most_common}")
```

```python
# 13. Написати функцію, яка б множила всі значення в колонці, які менше середнього значения в цій колонці, на 2, і ділила інші значення на 4.

def transform_column(col):
    mean = col.mean()
    return np.where(col < mean, col * 2, col / 4)
```

```python
# 14. Застосувати отриману функцію до 3-ї колонки. Функція має прийняти лише одну колонку і повернути колонку із зміненими значеннями

col3_mean = data[:, 2].mean()
print(f"Column 3 mean = {col3_mean}")
data[:, 2] = transform_column(data[:, 2])
print(f"Transformed data\n{data}")
```
