```python
import pandas as pd
from black.comments import make_comment
```

```python
data = pd.read_csv('adult.data.csv')
data.head()
```

**1. Скільки чоловіків и жінок (ознака *sex*) представлено в цьому наборі даних?**

```python
male_count = (data["sex"] == "Male").sum()
print(f"Male count: {male_count}")
female_count = (data["sex"] == "Female").sum()
print(f"Female count: {female_count}")
```

**2. Який середній вік (ознака *age*) жінок?**

```python
mean_age = data.loc[data["sex"] == "Female", "age"].mean()
print(f"Female mean age: {mean_age}")
```

**3. Яка доля громадян Німеччини (ознака *native-country*)?**

```python
german_citizens_ratio = (data["native-country"] == "Germany").mean()
print(f"German citizens ratio: {german_citizens_ratio}")
```

**4-5. Які средні значення та средньоквадратичні відхилення віку тих, хто отримує більше 50K на рік (ознака *salary*) і тих, хто отримує менше 50K на рік?**

```python
gt_50k_age = data.loc[data["salary"] == ">50K", "age"]
gt_50k_age.mean()
print(f"Mean of age for people with a salary >50K: {gt_50k_age.mean()}")
gt_50k_age.std()
print(f"Std of age for people with a salary >50K: {gt_50k_age.std()}")
lte_50k_age = data.loc[data["salary"] == "<=50K", "age"]
lte_50k_age.mean()
print(f"Mean of age for people with a salary <=50K: {lte_50k_age.mean()}")
lte_50k_age.std()
print(f"Std of age for people with a salary <=50K: {lte_50k_age.std()}")
```

**6. Чи правда, що люди, які отримують більше 50k, мають як мінімум вищу освіту? (ознака *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
allowed = ["Bachelors", "Prof-school", "Assoc-acdm", "Assoc-voc", "Masters", "Doctorate"]
all_have = data.loc[data["salary"] == ">50K", "education"].isin(allowed).all()
print(
    'All people with a salary >50K have higher education'
    if all_have
    else 'Not all people with a salary >50K have higher education'
)
```

**7. Порахуйте статистику віку для кожної раси (ознака *race*) і кожної статі. Використайте *groupby* и *describe*. Знайдіть таким сином максимальний вік чоловіків раси *Amer-Indian-Eskimo*.**

```python
aie_max_age = data.groupby("race")["age"].describe().loc["Amer-Indian-Eskimo", "max"]
print(f"Max age of Amer-Indian-Eskimo: {aie_max_age}")
```

**8. Серед кого більша доля багатіших (>50K): серед одружених чи не одружених чоловіків (ознака *marital-status*)? За одружених вважаємо тих, у кого *marital-status* починається із *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), всі інші - не одружені.**

```python
men = data[data["sex"] == "Male"]
married_mask = men["marital-status"].str.startswith("Married")
married_men = men[married_mask]
unmarried_men = men[~married_mask]
married_men_gt_50k_ratio = (married_men["salary"] == ">50K").mean()
print(f"Married men earn >50K ratio: {married_men_gt_50k_ratio}")
unmarried_men_gt_50k_ratio = (unmarried_men["salary"] == ">50K").mean()
print(f"Unmarried men earn >50K ratio: {unmarried_men_gt_50k_ratio}")
print(
    'Ratio of married men who earn >50K is bigger than unmarried'
    if married_men_gt_50k_ratio > unmarried_men_gt_50k_ratio
    else 'Ratio of unmarried men who earn >50K is bigger than married'
)
```
